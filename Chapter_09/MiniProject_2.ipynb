```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "white")
%matplotlib inline
```

```python
# Get Dataset from Drive
url = 'https://drive.google.com/uc?export=download&id=1DvZzmN8ASp3LnZ3dgj2x4Xq0a3ZKX_ua'
df = pd.read_csv(url,encoding = "latin-1")
df.sample(5)
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
sns.histplot(data=df, x='yearOfRegistration', bins=50, kde=True, stat='density')
plt.title('Distribuição de Veículos com Base no Ano de Registro')
plt.xlabel('Ano de Registro')
plt.ylabel('Densidade (KDE)')
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
sns.set(rc={'figure.figsize':(12,8)})
sns.boxplot(data=df, x='vehicleType', y='price')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Preço')
plt.title('Análise de Outliers')
plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
sns.catplot(data=df, x='vehicleType', kind='count',aspect=2, palette='colorblind')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Total de Veículos')
plt.title('Veículos à venda por tipo')
plt.show()
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.catplot(data=df, y='brand', kind='count',aspect=2, palette='colorblind')
plt.xlabel('Total de Veículos')
plt.ylabel('Marca')
plt.title('Veículos à venda por marca')
plt.show()
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
pricePerType = df.groupby(['vehicleType','gearbox']).mean()['price']
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='vehicleType', y='price', hue='gearbox', palette='colorblind')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Preço Médio')
plt.title('Preço médio de veículos por tipo')
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
powerPerType = df.groupby(['vehicleType','gearbox']).mean()['powerPS']
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='vehicleType', y='powerPS', hue='gearbox', palette='colorblind')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Potência Média')
plt.title('Potência média de veículos por tipo')
plt.show()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Calcule a média de preço por marca e por veículo
avgPricePerBrand = df.groupby(['brand','vehicleType']).mean().fillna(0)
avgPricePerBrand = avgPricePerBrand.reset_index()[['price','brand','vehicleType']].rename(columns={'price': 'avgPrice'})
avgPricePerBrand.head()
```

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
plt.figure(figsize=(15,25))
sns.heatmap(avgPricePerBrand.pivot('brand','vehicleType','avgPrice'), annot=True, fmt=".2f")
plt.xlabel('Tipo de Veículo')
plt.ylabel('Potência Média')
plt.title('Potência média de veículos por tipo')
plt.show()
```
